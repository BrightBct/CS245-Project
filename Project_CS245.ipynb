# Data Cleaning

**ส่วนที่ 2 Data Cleaning** 

ก่อนจะใช้งานข้อมูล ต้องจัดการข้อมูลให้มีความถูกต้องเรียบร้อยก่อน

เนื่องจากชุดข้อมูลที่หามาเป็นข้อมูลที่มีเพียงตัวเลข จึงได้ทำการเช็คว่าข้อมูลที่นำมาใช้นั้น มีค่า Nan ในชุดข้อมูลหรือไม่ ด้วยการใช้คำสั่ง dataset.isna() โดยถ้าใช้คำสั่งเพียงเท่านี้ จะได้ผลลัพธ์เป็น array True False ของข้อมูล ซึ่งไม่สามารถระบุได้ว่าข้อมูลส่วนไหนมีค่าเป็น Nan จึงใช้คำสั่ง sum() เพิ่มเข้าไป เพื่อที่จะได้นำค่า True False ของแต่ละ column มารวมกัน ซึ่งจะได้เป็นผลรวมของค่า True False แต่ละ column ถ้า column ไหนไม่ได้มีค่าเป็น 0 แสดงว่า column นั้น มีค่า Nan อยู่

```
print("Dataset Nan: ")
print(dataset.isna().sum())
```

ผลลัพธ์ของคำสั่ง dataset.isna().sum() ทุกค่าของแต่ละ column เป็น 0 หมด จึงไม่ต้องมีการจัดการกับค่า Nan

ทำการเช็คว่าใน dataset มีค่าที่ซ้ำกันไหม ด้วยการใช้ dataset.duplicated() โดยถ้าใช้คำสั่งเพียงเท่านี้ จะได้ผลลัพธ์เป็น array True False ของข้อมูล ซึ่งไม่สามารถระบุได้ว่าข้อมูลส่วนไหนซ้ำกัน จึงใช้คำสั่ง sum() เพิ่มเข้าไป เพื่อที่จะได้นำค่า True False ของแต่ละ row มารวมกัน ซึ่งจะได้เป็นผลรวมของค่า True False ทุก row ถ้าค่าที่ออกมาไม่ได้มีค่าเป็น 0 แสดงว่าข้อมูลชุดนี้มีค่าซ้ำกันอยู่

```
print("Dataset duplicated: ",dataset.duplicated().sum())
```

ผลลัพธ์ของคำสั่ง dataset.duplicated().sum() มีค่าเป็น 1 แสดงว่าในชุดข้อมูลมีแถวที่ซ้ำอยู่ 1 แถว จึงต้องมีการจัดการกับข้อมูลที่ซ้ำกัน

```
duplicated = dataset[dataset.duplicated()]
display(duplicated)
```

เนื่องจากข้อมูลนี้มีแถวที่ซ้ำอยู่ 1 แถว จึงได้ทำการ drop แถวที่ซ้ำ โดยการใช้คำสั่ง dataset.drop_duplicates(inplace=True) แล้วลองเรียกใช้ dataset.duplicated().sum() อีกรอบเพื่อเช็คว่าได้ทำการ drop แถวที่ซ้ำไปแล้วหรือยัง


```
dataset.drop_duplicates(inplace=True)
print("Dataset duplicated: ",dataset.duplicated().sum())
```

เรียกใช้คำสั่งเดิมเพื่อดูว่าหลังจาก drop แถวทิ้งไปแล้ว ชุดข้อมูลจะเหลือขนาดเท่าไหร่ แล้วจะเหลือกี่จำนวน

```
print("dataset shape:", dataset.shape)
print("dataset size: ", dataset.size)
```

# Data Wrangling

**ส่วนที่ 3 Data Wrangling** 

หลังจากทำ Data Cleaning เสร็จแล้ว ก็จะต้องเตรียมข้อมูลให้พร้อมใช้งานสำหรับส่วนอื่น ๆ

ทำการจัดกลุ่มของข้อมูลตาม Type ด้วยคำสั่ง dataset.groupby(‘Type’) เก็บไว้ในตัวแปร group หลังจากนั้นแสดงผลลัพธ์ โดยถ้าใช้คำสั่งเพียงเท่านี้ ผลลัพธ์จะออกมาเป็น pandas.core.groupby.generic.DataFrameGroupBy ซึ่งหมายถึงข้อมูลที่ยังไม่ได้ถูกนำมาใช้ แค่จัดกลุ่มข้อมูลไว้เท่านั้น จึงใช้คำสั่ง size() ต่อ เพื่อจะได้ดูว่าหลังจากจัดกลุ่มตาม Type แล้ว แต่ละ Type จะมีข้อมูลอยู่กี่จำนวน

```
group = dataset.groupby('Type')
print(group.size())
```

- กระจกประเภทที่ 5 อยู่จำนวน 13 ข้อมูล

- กระจกประเภทที่ 6 อยู่จำนวน 9 ข้อมูล

- กระจกประเภทที่ 7 อยู่จำนวน 29 ข้อมูล

รวมทั้งสิ้น 213 ข้อมูล

```
display(group.get_group(3))
```

**ส่วนข้างล่างนี้ทำเพื่อจัดเตรียมข้อมูลไว้ใช้ในหัวข้อถัดไป**

ทำการจัดเก็บข้อมูลของแต่ละ Type ลงในตัวแปรที่แสดงถึงชื่อ Type นั้น ๆ เช่น จัดเก็บ Type ที่ 1 ลงในตัวแปร Type1 เป็นต้น โดยได้ drop column Type ไว้ เพราะแต่ละตัวแปร หมายถึงแต่ละ Type จึงไม่จำเป็นต้องมี column Type ที่ระบุว่าข้อมูลนี้อยู่ Type ไหน

```
Type1 = group.get_group(1).drop(['Type'], axis=1)
Type2 = group.get_group(2).drop(['Type'], axis=1)
Type3 = group.get_group(3).drop(['Type'], axis=1)
Type5 = group.get_group(5).drop(['Type'], axis=1)
Type6 = group.get_group(6).drop(['Type'], axis=1)
Type7 = group.get_group(7).drop(['Type'], axis=1)
```

ทำการสร้างตัวแปรที่เก็บชื่อของ column เอาไว้ โดยสร้าง 2 ตัวแปร 

ตัวแปรที่ 1 : col_groupType คือตัวแปรที่เก็บชื่อของ column ตัวแปรนี้ไม่มี column 'Type'

ตัวแปรที่ 2 : col_dataset คือตัวแปรที่เก็บชื่อของ column ตัวแปรนี้มี column 'Type'

```
col_groupType = []
col_dataset = []
for col in dataset.columns:
  col_dataset.append(col)
  if(col == 'Type'):
    continue
  col_groupType.append(col)
print('col_groupType:\t', col_groupType)
print('col_dataset:\t', col_dataset)
```

ทำการ import train_test_split มาใช้สำหรับแยกข้อมูล เพื่อแยกส่วนข้อมูลเป็น train กับ validation ในการทำ Machine Learning

```
from sklearn.model_selection import train_test_split
```

ทำการแปลงข้อมูลใน dataset ไปเก็บใน array 

หลังจากนั้นนำข้อมูลที่เป็นแร่ธาตุ (column ที่เป็นแร่ธาตุ) เก็บเอาไว้ในตัวแปร X และเก็บข้อมูลที่บอกว่าเป็นกระจกประเภทอะไร (column Type) ไว้ในตัวแปร Y 

จากนั้นทำการแยก X เป็น X_train กับ X_validation และแยก Y เป็น Y_train กับ Y_validation ด้วยคำสั่ง train_test_split(X, Y, test_size = 0.2, random_state=1)

โดยการทำ train_test_split สามารถเลือกได้ว่าจะแยกข้อมูลเท่าไหร่ ในทีนี้จะแยกข้อมูลเป็น 80% (170 แถว) กับ 20% (43 แถว) ซึ่ง 80% ไปเก็บไว้ใน X_train กับ Y_train และอีก 20% ไปเก็บไว้ใน X_validation กับ Y_validation และกำหนดให้ random_state = 1 เพื่อเวลาสุ่มอีกรอบ จะได้ข้อมูลเดิม

สิ่งนี้ทำเพื่อให้ระบบคอมพิวเตอร์เรียนรู้จาก train แล้วนำสิ่งที่ได้เรียนรู้จาก train ไปคาดเดา validation

```
array = dataset.values
X = array[:,0:9]
Y = array[:,9]
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=0.2, random_state=5)
```

```
print('X_train:')
print(X_train)
print()
print('X_validation:')
print(X_validation)
print()
print('Y_train:')
print(Y_train)
print()
print('Y_validation:')
print(Y_validation)
print()
```

Y_train กับ Y_validation เป็นข้อมูลที่บอกว่าเป็นกระจกชนิดไหน ซึ่งแต่เดิมเป็นข้อมูลประเภท int หลังจากแยกข้อมูลทำให้กลายเป็นข้อมูลประเภท float ดังนั้นเพื่อความมั่นใจว่าข้อมูลจะไม่ผิดพลาด จึงทำการเปลี่ยนข้อมูลจาก float ให้กลายเป็น int

```
print('X_validation:')
print(X_validation)
print()
print('Y_train:')
print(Y_train)
print()
print('Y_validation:')
print(Y_validation)
print()
```

ทำการ import cross_val_score กับ StratifiedKFold เพื่อมาใช้ในการเลือก model ในการทำ Machine Learning 

```
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

ทำการ import classification_report, confusion_matrix และ accuracy_score เพื่อนำมาหาความแม่นยำของแต่ละ model

```
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

ทำการ import แต่ละ method เพื่อนำมาทำ machine learning โดยเลือกเฉพาะ method ที่สามารถทำ Classification ได้

```
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

ทำการสร้างตัวแปร model เพื่อเก็บแต่ละ method 

```
models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC(gamma='auto')))
```

ทำการเข้าลูป เพื่อหาความแม่นยำของแต่ละ method โดยใช้ StratifiedKFlod และ cross_val_score ซึ่งกำหนดการคำนวณความแม่นยำด้วย accuracy ซึ่งผลลัพธ์ที่ได้ออกมานั้นคือการหา accuracy_score ของแต่ละ method ซึ่งผลลัพธ์ที่ได้นั้น ถ้า method ไหนมีค่าเข้าใกล้ 1 มากที่สุด แสดงว่า method นั้นมีความแม่นยำสูง หลังจากนั้นจะทำการเลือก method นั้นในการทำ machine learning ต่อไป

```
results = []
names = []
for name, model in models:
	kfold = StratifiedKFold(n_splits=6, random_state=1, shuffle=True)
	cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
	results.append(cv_results)
	names.append(name)
	print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```

ผลลัพธ์ที่ได้คือ method CART (Decision Trees) มีค่าเข้าใกล้ 1 มากที่สุด จึงเลือก method CART (Decision Trees) ในการทำ machine learning ต่อไป

ทำ bloxplot เพื่อเปรียบเทียบความแม่นยำของแต่ละ method โดยจะเห็นว่า method CART (Decision Trees) จะเข้าใกล้เลข 1 มากกว่า method อื่น ๆ

```
plt.figure(figsize=(10, 10))
plt.boxplot(results, labels=names)
plt.title('Algorithm Comparison');
```

ให้ model เป็น method CART (Decision Tree) แล้วทำการ fit(X_train, Y_train) ซึ่ง fit คือการให้ข้อมูลกับระบบคอมพิวเตอร์ โดยดูข้อมูลจาก X_train และผลลัพธ์จาก Y_train หลังจากนั้นให้คอมพิวเตอร์คาดเดาข้อมูลของ X_validation ด้วยคำสั่ง model.predict(X_validation)

```
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
predictions = model.predict(X_validation)
print(predictions)
```

ทำการเปรียบเทียบ prediction ที่ระบบคอมพิวเตอร์คาดเดา กับ Y_validation (คำตอบ) 

```
print('Accuracy Score: ',accuracy_score(Y_validation, predictions))
print()
print('Confusion Matrix')
print(confusion_matrix(Y_validation, predictions))
print()
print('Classification Report')
print(classification_report(Y_validation, predictions))
```
